```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import rotate
```

```python
def frequency_demo():
    sines = np.sin(np.linspace(0, 20*np.pi, 256))
    waves = np.outer(sines, sines)
    normal = rotate(waves, 0, mode='edge')[100:200, 100:200]
    rotated = rotate(waves, 45, mode='edge')[100:200, 100:200]
    f_normal = np.fft.fftshift(np.fft.fft2(normal))
    f_rotated = np.fft.fftshift(np.fft.fft2(rotated))
    plt.gray()
    plt.subplots(3, 2, figsize=(15, 20))
    plt.subplot(321)
    plt.imshow(normal)
    plt.subplot(322)
    plt.imshow(rotated)
    plt.subplot(323)
    plt.imshow(np.abs(f_normal))
    plt.subplot(324)
    plt.imshow(np.abs(f_rotated))
    plt.subplot(325)
    plt.imshow(np.fft.fftshift(np.exp(1j*np.angle(np.fft.fft2(normal)))).real)
    plt.subplot(326)
    plt.imshow(np.fft.fftshift(np.exp(1j*np.angle(np.fft.fft2(rotated)))).real)
    plt.show()

frequency_demo()
```
